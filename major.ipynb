```python
# 다전공 신청 현황 시각화를 위한 데이터 준비
import matplotlib.pyplot as plt
import seaborn as sns

# 다전공별 학생 수 계산
major_counts = df['다전공명'].value_counts()

# 시각화
plt.figure(figsize=(12, 6))
sns.barplot(x=major_counts.values, y=major_counts.index)
plt.title('2017~2025 정보보안암호수학과 학생들의 다전공 신청 현황')
plt.xlabel('학생 수')
plt.ylabel('다전공명')

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'

plt.tight_layout()
plt.show()

# 원본 데이터프레임은 유지
df = df[['다전공명', '학년', '학적\n상태']]
```
